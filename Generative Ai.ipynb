```python
## Question 1: What is Generative AI and what are its primary use cases across industries?

Answer: Generative AI is a type of AI that can create new content like:
• text

• images

• music

• videos

• code

It learns from data and then makes new similar data.

Its primary use cases across industries are :

 1. Education

• write notes


• explain topics

• make questions and answers

2. Healthcare

• help doctors write reports

• make medical images

• help in drug discovery

 3. Business

• write emails


• make ads

• customer support chatbots

## Question 2: Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?

Answer : Role of probabilistic modeling in generative models

Probabilistic modeling means using probability to describe data.

In generative models:
• They learn the probability of data


• They learn how likely something is to appear

• They can create new data (text, image, sound)

Example (idea):
The model learns:

P(data) or P(data | class)

So it can make new similar data

Difference between Generative and Discriminative models

* Generative models


    * Learn how data is made
    * Learn full data pattern• Can generate new data

Example task:
“Make a new sentence like this.”

* Discriminative models

   *  Learn difference between classes
   *  Learn only decision boundary
    * Used for classification

Example task:
“Is this email spam or not?

## Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

Answer: Autoencoders (AE)

• Learn to compress and reconstruct text


• Latent space is fixed (one point)

• Mostly used for reconstruction, not good for new text

• Output is usually same as input

 Example use:
Rebuild the same sentence:
“The sky is blue” - “The sky is blue”

Variational Autoencoders (VAE)

• Learn probability distribution of text


• Latent space is smooth (range of values)

• Can generate new similar sentences

• Better for text generation

 Example use:
“The sky is blue” -
“The night is dark” (new but similar style)

## Question 4: Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?


Answer: The working of attention mechanisms in Neural Machine Translation (NMT) are :

* The encoder reads the source sentence

Example:
I love you

* The decoder generates the target sentence

Example:
Te amo

* At each word, attention checks all encoder words

 It asks:
“Which source word is most important now?”

* It gives higher weight to important words

Example:
While making amo, it focuses more on love.

* The decoder uses this focused information to produce the next word.

## Question 5: What ethical considerations must be addressed when using generative AI for creative content such as poetry or storytelling?

Answer: Ethical issues
1. Plagiarism (copying)

AI should not copy poems or stories from real writers.
It must create new content.

2. Copyright

AI should not use protected books or poems without permission.
Writers’ work must be respected.

3. Bias and harmful content

AI may write content that is:

• unfair

• racist

• offensive

This must be controlled and checked.

4. Fake authorship

People should know when a poem or story is written by AI.
Do not claim: “I wrote this” if AI made it.

5. Job impact

AI can affect writers and artists’ jobs.
It should help humans, not fully replace them.

6. Misuse

AI stories can be used to spread:
• lies

• hate

• wrong ideas

So rules are needed.
```

```python
# Question 6: Use the following small text dataset to train a simple Variational
# Autoencoder (VAE) for text reconstruction:
# ["The sky is blue", "The sun is bright", "The grass is green",
# "The night is dark", "The stars are shining"]
# 1. Preprocess the data (tokenize and pad the sequences).
# 2. Build a basic VAE model for text reconstruction.
# 3. Train the model and show how it reconstructs or generates similar sentences.
# Include your code, explanation, and sample outputs

sentences = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]


import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)
max_len = max(len(seq) for seq in sequences)

padded = pad_sequences(sequences, maxlen=max_len, padding="post")

vocab_size = len(tokenizer.word_index) + 1

print("Padded sequences:")
print(padded)



import tensorflow as tf
from tensorflow.keras import layers, Model

latent_dim = 2

# Encoder
inputs = layers.Input(shape=(max_len,))
x = layers.Embedding(vocab_size, 16)(inputs)
x = layers.Flatten()(x)
h = layers.Dense(32, activation="relu")(x)

z_mean = layers.Dense(latent_dim)(h)
z_log_var = layers.Dense(latent_dim)(h)

def sampling(args):
    z_mean, z_log_var = args
    eps = tf.random.normal(shape=(tf.shape(z_mean)[0], latent_dim))
    return z_mean + tf.exp(0.5 * z_log_var) * eps

z = layers.Lambda(sampling)([z_mean, z_log_var])

# Decoder
decoder_h = layers.Dense(32, activation="relu")(z)
decoder_out = layers.Dense(max_len * vocab_size, activation="softmax")(decoder_h)
outputs = layers.Reshape((max_len, vocab_size))(decoder_out)

vae = Model(inputs, outputs)

vae.compile(optimizer="adam", loss="sparse_categorical_crossentropy")
vae.summary()
```

```python
# Question 8: Implement a simple attention-based encoder-decoder model for
# English-to-Spanish translation using Tensorflow or PyTorch.

import tensorflow as tf
import numpy as np
! pip install tensorflow
# Dataset
eng = ["i am a student", "i love you", "how are you"]
spa = ["yo soy estudiante", "te amo", "como estas"]

# Tokenization
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

eng_tok = Tokenizer()
spa_tok = Tokenizer()

eng_tok.fit_on_texts(eng)
spa_tok.fit_on_texts(spa)

eng_seq = eng_tok.texts_to_sequences(eng)
spa_seq = spa_tok.texts_to_sequences(spa)

max_len = 5

eng_pad = pad_sequences(eng_seq, maxlen=max_len, padding="post")
spa_pad = pad_sequences(spa_seq, maxlen=max_len, padding="post")

eng_vocab = len(eng_tok.word_index) + 1
spa_vocab = len(spa_tok.word_index) + 1

# Encoder
class Encoder(tf.keras.Model):
    def __init__(self, vocab, embed_dim, units):
        super().__init__()
        self.embedding = tf.keras.layers.Embedding(vocab, embed_dim)
        self.gru = tf.keras.layers.GRU(units, return_sequences=True, return_state=True)

    def call(self, x):
        x = self.embedding(x)
        output, state = self.gru(x)
        return output, state

# Attention Layer
class Attention(tf.keras.layers.Layer):
    def __init__(self, units):
        super().__init__()
        self.W1 = tf.keras.layers.Dense(units)
        self.W2 = tf.keras.layers.Dense(units)
        self.V = tf.keras.layers.Dense(1)

    def call(self, enc_out, hidden):
        hidden = tf.expand_dims(hidden, 1)
        score = self.V(tf.nn.tanh(self.W1(enc_out) + self.W2(hidden)))
        weights = tf.nn.softmax(score, axis=1)
        context = tf.reduce_sum(weights * enc_out, axis=1)
        return context

# Decoder
class Decoder(tf.keras.Model):
    def __init__(self, vocab, embed_dim, units):
        super().__init__()
        self.embedding = tf.keras.layers.Embedding(vocab, embed_dim)
        self.gru = tf.keras.layers.GRU(units, return_sequences=True, return_state=True)
        self.fc = tf.keras.layers.Dense(vocab)
        self.att = Attention(units)

    def call(self, x, hidden, enc_out):
        context = self.att(enc_out, hidden)
        x = self.embedding(x)
        x = tf.concat([tf.expand_dims(context,1), x], axis=-1)
        out, state = self.gru(x)
        out = self.fc(out)
        return out, state

# Model parameters
embedding_dim = 16
units = 32

encoder = Encoder(eng_vocab, embedding_dim, units)
decoder = Decoder(spa_vocab, embedding_dim, units)

# Forward pass (test run)
sample_input = tf.convert_to_tensor(eng_pad)

enc_out, enc_hidden = encoder(sample_input)

dec_input = tf.expand_dims([1,1,1], 1)  # start token simulation

pred, dec_hidden = decoder(dec_input, enc_hidden, enc_out)

print("Encoder output shape:", enc_out.shape)
print("Decoder prediction shape:", pred.shape)
```
